(a) Write a program in Python or Java that uses the OMDb API to lookup/search _your_ top 10 choices of movie titles and extract their imdbIDs. Print the movie titles along with their imdbID to the screen. (E.g., search for "Dangal" or "Wolf Warrior 2" and record the imdbID that corresponds to the exact movie you have in mind).

```python
#creating an array of top 10 movies
top_choice = ['The Truman Show', 'The pursuit of happyness', 'Life is beautiful', 'Toy story', 'The Classic', 
              'The Lord of the Rings: the two towers', 'Flipped', 'The Lord of the Rings: the return of the king', 
              'Avatar', 'Doctor Strange']
```

```python
print(top_choice)
```

```python
#Creating an empty list to store movie IDs
top_choice_id = []
```

```python
#running the loop to parse all movie details
```

```python
for i in range(len(top_choice)):
    url_top = "http://omdbapi.com/?apikey=dcd209b4&t="+top_choice[i]
    response_top = requests.get(url_top)
    data_top = response_top.text 
    parsed_top = json.loads(data_top)
    print(json.dumps(parsed_top, indent=4))
    top_choice_id.append(parsed_top['imdbID'])
```

```python
top_choice_id
```

```python
import pandas as pd
```

```python
print(pd.DataFrame({'Title':top_choice, 'imdbID':top_choice_id}))
```

```python
for i in range(len(top_choice_id)):
    url_top_id = "http://omdbapi.com/?apikey=dcd209b4&i="+top_choice_id[i]
    response_top_id = requests.get(url_top_id)
    data_top_id = response_top.text 
    parsed_top_id = json.loads(data_top_id)
    print(json.dumps(parsed_top_id, indent=4))
    print("-----------------------------------------------------------------")
```
